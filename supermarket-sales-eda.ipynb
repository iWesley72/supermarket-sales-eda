## Análise exploratória

#### Lucro x Vendas

```python
table = pd.pivot_table(dfSales, index='Month', values=['Invoice ID', 'gross income'], aggfunc={'Invoice ID':'count', 'gross income':'sum'})
table
```

```python
fig = px.line(table, x=table.index.get_level_values('Month'), y=table.columns, title='Pedidos x Lucro bruto')
fig.update_xaxes(title='Mês')
fig.update_yaxes(title='')
fig.show()
```

```python
fig = px.histogram(table, barmode='group', x=table.index.get_level_values('Month'), y=table.columns, histfunc='sum', title='Crescimento lucro bruto x pedidos')
fig.update_layout(bargap=0.5)
fig.update_xaxes(title='Mês')
fig.update_yaxes(title='')
fig.show()
```

```python
fig = px.pie(table, names=table.index.get_level_values('Month'), values='gross income', title='Percentual de lucro bruto por mês')
fig.show()
```

#### a) Lucro e pedidos

```python
dfSales.head()
```

```python
table = pd.pivot_table(dfSales, index=['City', 'Month'], values=['Invoice ID', 'gross income'], aggfunc={'Invoice ID':'count', 'gross income':'sum'})
table
```

Todas as unidades são bem próximas em número de pedidos, vamos plotar um gráfico de barra para entender o crescimento, ou queda, de lucro bruto e vendas no periodo estudado

```python
fig = px.histogram(table, x=table.index.get_level_values('Month'), y='gross income', color=table.index.get_level_values('City'), title='Crescimento do lucro bruto por cidade')
fig.update_layout(bargap=0.5)
fig.update_xaxes(title='Mês')
fig.update_yaxes(title='Lucro bruto')
fig.show()
```

```python
fig = px.histogram(table, x=table.index.get_level_values('Month'), y='Invoice ID', color=table.index.get_level_values('City'), title='Crescimento de pedidos por cidade')
fig.update_layout(bargap=0.5)
fig.update_xaxes(title='Mês')
fig.update_yaxes(title='Qtd. pedidos')
fig.show()
```

#### c) Membro x Comum

Temos 501 clientes membros para 499 clientes comuns. Relação de aproximadamente 1:1. Vamos ver se houve aumento de membros no decorrer do tempo.

Levando em consideração a distribuição de membros por cidade e o crescimento de membros no periodo de três meses, podemos conculuir que a proporção é de 1:1, e que ou não é tão vantajoso ser membro ou não é tão divulgado como deveria.
O número de membros por unidade e mês se mantem praticamente o mesmo, com pouca variação.

#### d) Linha de produto

```python
dfSales.head()
```

```python
table = pd.pivot_table(dfSales, index='Product line', values='Invoice ID', aggfunc='count')
table
```

```python
fig = px.pie(table, names=table.index.get_level_values('Product line'), values='Invoice ID', title='Percentual de pedido por linha de produto')
fig.show()
```

É bem distribuida a quantidade de pedidos, onde quase todos tem o mesmo percentual com relação ao total de pedidos. De uma visão geral, não tem uma linha que o público tem maior preferência. Vamos ver com relação ao lucro bruto.

```python
table = pd.pivot_table(dfSales, index='Product line', values='gross income', aggfunc='sum')
table
```

Aparentemente, a proporção se mantém para o lucro bruto, onde todas as linhas tem um valor muito aproximado.

```python
fig = px.pie(table, names=table.index.get_level_values('Product line'), values='gross income', title='Percentual de lucro por linha de produto')
fig.show()
```

Como comentado, a proporção se manteve com relação ao percentual, mas mudou na relação lucro/pedidos. Onde acessórios de moda que representava o maior número de pedidos teve o quarto maior lucro bruto. Sendo comidas e bebidas que obteve o maior lucro bruto, mas a diferença monetária entre os dois foi de apenas 100,00.

Vamos ver se existe então alguma relação com gênero.

```python
fig = px.histogram(dfSales, x='Product line', y='Invoice ID', color='Gender', histfunc='count', title='Linha de produto x gênero (pedidos)', barmode='group')
fig.update_layout(bargap=0.5)
fig.update_xaxes(title='Linha de produto')
fig.update_yaxes(title='Qtd. de pedidos')
fig.show()
```

Gênero feminino tem clara preferência aos produtos de esportes e viagem e acessórios de moda. Enquanto o gênero masculino tem preferência aos produtos de saúde e beleza.

Vamos ver se existe alguma preferência por unidade de mercado.

```python
table = pd.pivot_table(dfSales, index=['City', 'Product line'], values='Invoice ID', aggfunc='count')
table
```

```python
fig = px.histogram(table, x=table.index.get_level_values('Product line'), y='Invoice ID', color=table.index.get_level_values('City'), barmode='group', title='Linha de produto por cidade (pedidos)')
fig.update_layout(bargap=0.5)
fig.update_xaxes(title='Linha de produto')
fig.update_yaxes(title='Qtd. de pedidos')
fig.show()
```

  Os principais produtos vendidos em Mandalay são de esportes e viagem e acessórios de moda.
  Em Naypyitaw são produtos alimenticios e acessórios de moda.
  Já em Yangon são casa e estilo de vida e eletrônicos.
  Os produtos de saúde e beleza são vendidas de mesma quantia em Mandalay e Naypyitaw, e em menor quantia em Yangon.

#### e) Método de pagamento

```python
dfSales.head()
```

```python
fig = px.histogram(dfSales, x='Payment', title='Método de pagamento')
fig.update_xaxes(title='Método de pagamento')
fig.update_yaxes(title='Frequência')
fig.update_layout(bargap=0.5)
fig.show()
```

De forma geral, o método mais comum é e-wallet e dinheiro, mesma frequência, enquanto cartão crédito é menos frequente.

```python
fig = px.histogram(dfSales, x='City', color='Payment', barmode='group', title='Método de pagamento preferido por cidade')
fig.update_layout(bargap=0.5)
fig.update_xaxes(title='Cidade')
fig.update_yaxes(title='Frequência')
fig.show()
```

Em Yangon a preferência é claramente e-wallet, em Naypyitaw dinheiro e em Mandalay não há método de pagamento preferido.

```python
fig = px.histogram(dfSales, x='Payment', color='Gender', barmode='group', title='Método de pagamento preferido por gênero')
fig.update_layout(bargap=0.5)
fig.update_xaxes(title='Método')
fig.update_yaxes(title='Frequência')
fig.show()
```

Método preferido pelo gênero feminino foi dinheiro, enquanto masculo e-wallet. Por curiosidade, o pagamento por cartão de crédito é preferido pelo gênero femino.

## Conclusões

*   O maior lucro foi no mês de janeiro, lucro de 5537.71, com aproximadamente 352 pedidos.
*   Houve uma queda no lucro de 5.9% comparando janeiro e março.
*   A cidade com o maior lucro é Naypyitaw, aproximadamente 5265.18.
*   As cidades com a maior pontuação, satisfação do cliente, são Naypyitaw e Yangon, com 7.07 e 7.02 respectivamente.
*   A proporção entre membros e clientes é de 1:1.
*   A linha produtos com maior lucro é alimenticio, com lucro de 2673.56.
*   O produto mais coomprado pelo gênero feminino é acessórios de moda, enquanto do gênero masculino são produtos de saúde e beleza.
*   A loja de Yangon é a que mais vende produtos de casa e estilo de vida.
*   Naypyitaw é a que mais vende produtos alimenticios.
*   Mandalay a que mais vende produtos esportivos e de viagem.
*   Os métodos de pagamento mais recorrentes são dinheiro e e-wallet, com a mesma frequência.
*   Em Yangon é mais comum e-wallet e em Naypyitaw dinheiro. Não há preferência em Mandalay.
*   O gênero feminino tende a pagar mais em dinheiro e o masculino com e-wallet.
*   A proporção de clientes do gênero masculino para o gênero feminino é de 1:1.
